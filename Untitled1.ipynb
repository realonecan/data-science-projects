saldo

This file contains info about account balances and transactions.

id ‚Äì account id (used to connect with other tables)

v_date ‚Äì date of the operation

s_in ‚Äì incoming balance at the start of the day

s_out ‚Äì outgoing balance at the end of the day

dt ‚Äì debit turnover (how much was debited)

ct ‚Äì credit turnover (how much was credited)

bs_in ‚Äì balance at the beginning of the reporting period

bs_out ‚Äì balance at the end of the reporting period

bct ‚Äì total transaction difference

date_next_oper ‚Äì next operation date if it exists

bdt ‚Äì cumulative debit until the next operation

account

Contains basic account info.

id ‚Äì same as in saldo, identifies the account

client ‚Äì client name or code

db_test_1

Client personal details.

id ‚Äì client id

birthday ‚Äì date of birth

birth_place ‚Äì place of birth

code_gender ‚Äì gender code (1 = male, 2 = female or similar)

firebase_1 / firebase_2

These are event logs from Firebase, basically user activity inside the app.

id ‚Äì user or client id

platform ‚Äì android, ios, web, etc.

event_date ‚Äì when the event happened

event_name ‚Äì type of event (login, purchase, etc.)

category ‚Äì category of the event

language ‚Äì user interface language

mobile_brand_name ‚Äì phone brand

category_id ‚Äì numeric version of category

device_id1 ‚Äì device id from Firebase

engagement_time_msec ‚Äì how long user was active (in ms)

engaged_session_event ‚Äì 1 if it was an active session, else 0

```python
# ============================================================================
# STEP 9: MISSING VALUES ANALYSIS
# ============================================================================
print("\n" + "="*80)
print("STEP 9: MISSING VALUES ANALYSIS")
print("="*80)

# Calculate missing values
missing_count = df_full.isnull().sum()
missing_percent = (missing_count / len(df_full)) * 100

# Create a summary dataframe
missing_summary = pd.DataFrame({
    'Column': missing_count.index,
    'Missing_Count': missing_count.values,
    'Missing_Percent': missing_percent.values
})

# Show only columns with missing values
missing_summary = missing_summary[missing_summary['Missing_Count'] > 0].sort_values(
    'Missing_Percent', ascending=False
)

if len(missing_summary) > 0:
    print("\n‚ö†Ô∏è COLUMNS WITH MISSING VALUES:")
    print(missing_summary.to_string(index=False))
else:
    print("\n‚úÖ NO MISSING VALUES - Perfect!")

print(f"\nüìä SUMMARY:")
print(f"Total columns: {len(df_full.columns)}")
print(f"Columns with missing data: {len(missing_summary)}")
print(f"Total rows: {len(df_full):,}")
```

```python
# ============================================================================
# STEP 10: DATA CLEANING (SIMPLE FIX)
# ============================================================================
print("\n" + "="*80)
print("STEP 10: DATA CLEANING & PREPARATION")
print("="*80)

# 10.1: Handle date_next_oper missing values (SIMPLE approach)
print("\n1Ô∏è‚É£ Handling missing date_next_oper...")
print(f"   Missing values before: {df_full['date_next_oper'].isna().sum()}")

# Convert to string, fill, then back to datetime
df_full['date_next_oper'] = df_full['date_next_oper'].astype(str)
df_full.loc[df_full['date_next_oper'] == 'NaT', 'date_next_oper'] = '2099-12-31'
df_full['date_next_oper'] = pd.to_datetime(df_full['date_next_oper'], errors='coerce')

print(f"   Missing values after: {df_full['date_next_oper'].isna().sum()}")
print(f"   ‚úì Fixed!")

# 10.2: Check for duplicates
print("\n2Ô∏è‚É£ Checking duplicates...")
duplicates = df_full.duplicated().sum()
print(f"   Duplicate rows: {duplicates}")
if duplicates > 0:
    print(f"   Removing duplicates...")
    df_full = df_full.drop_duplicates()
    print(f"   ‚úì Removed! New shape: {df_full.shape}")

# 10.3: Check for outliers
print("\n3Ô∏è‚É£ Outlier check in key numeric columns...")
numeric_cols = ['s_in', 's_out', 'bdt', 'dt', 'ct']

for col in numeric_cols:
    Q1 = df_full[col].quantile(0.25)
    Q3 = df_full[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df_full[col] < (Q1 - 1.5 * IQR)) | (df_full[col] > (Q3 + 1.5 * IQR))).sum()
    outlier_pct = (outliers / len(df_full)) * 100
    print(f"   {col}: {outliers} outliers ({outlier_pct:.1f}%)")

print("\n4Ô∏è‚É£ Final cleaned dataset info:")
print(f"   Shape: {df_full.shape}")
print(f"   Unique clients: {df_full['client'].nunique()}")

print("\n‚úÖ DATA CLEANING COMPLETE!")
```

TARGETT!!

```python
# ============================================================================
# STEP 11: CREATE CHURN TARGET (FIXED)
# ============================================================================
print("\n" + "="*80)
print("STEP 11: CREATING CHURN TARGET")
print("="*80)

# Convert v_date to proper datetime without timezone
df_full['v_date'] = pd.to_datetime(df_full['v_date']).dt.tz_localize(None)

# Define time periods
observation_end = pd.Timestamp('2025-01-05')
prediction_start = pd.Timestamp('2025-01-06')

print(f"\nüìÖ TIME SPLIT:")
print(f"   Observation period: Jan 3-5 (training features)")
print(f"   Prediction period: Jan 6-8 (target outcome)")

# Split data by period
df_observation = df_full[df_full['v_date'] <= observation_end]
df_prediction = df_full[df_full['v_date'] >= prediction_start]

print(f"\nüìä DATA SPLIT:")
print(f"   Transactions in observation: {len(df_observation):,}")
print(f"   Transactions in prediction: {len(df_prediction):,}")

# Identify active clients in each period
clients_observation = set(df_observation['client'].unique())
clients_prediction = set(df_prediction['client'].unique())

print(f"\nüë• CLIENT ACTIVITY:")
print(f"   Active in observation period: {len(clients_observation):,}")
print(f"   Active in prediction period: {len(clients_prediction):,}")

# Define churn
churned_clients = clients_observation - clients_prediction  # Active then stopped
retained_clients = clients_observation & clients_prediction  # Active in both

print(f"\nüéØ TARGET DEFINITION:")
print(f"   CHURNED (1): Active Jan 3-5, NOT active Jan 6-8 = {len(churned_clients):,}")
print(f"   RETAINED (0): Active in both periods = {len(retained_clients):,}")
print(f"   Churn Rate: {len(churned_clients)/len(clients_observation)*100:.2f}%")

# Create target mapping
target_map = {}
for client in churned_clients:
    target_map[client] = 1  # Churned
for client in retained_clients:
    target_map[client] = 0  # Retained

# Create client-level dataset with target
client_target = pd.DataFrame({
    'client': list(target_map.keys()),
    'churn': list(target_map.values())
})

print(f"\n‚úÖ TARGET CREATED!")
print(f"   Total clients with target: {len(client_target):,}")
print(f"\nüìä Class Distribution:")
print(client_target['churn'].value_counts())
print(f"\nClass balance:")
print(client_target['churn'].value_counts(normalize=True).round(3)*100)
```

feature engeenering 

```python
# ============================================================================
# STEP 12: FEATURE ENGINEERING
# ============================================================================
print("\n" + "="*80)
print("STEP 12: FEATURE ENGINEERING")
print("="*80)

# Use ONLY observation period for features
df_features = df_observation.copy()

print(f"Creating features from {len(df_features):,} transactions...")

# ============================================================================
# GROUP 1: TRANSACTION FREQUENCY FEATURES
# ============================================================================
print("\n1Ô∏è‚É£ Transaction Frequency Features...")
transaction_freq = df_features.groupby('client').agg(
    total_transactions=('id_trans', 'count'),
    active_days=('v_date', 'nunique')
).reset_index()

# Calculate average transactions per day
transaction_freq['avg_trans_per_day'] = (
    transaction_freq['total_transactions'] / transaction_freq['active_days']
)

print(f"   ‚úì Created: total_transactions, active_days, avg_trans_per_day")

# ============================================================================
# GROUP 2: FINANCIAL BEHAVIOR FEATURES
# ============================================================================
print("\n2Ô∏è‚É£ Financial Behavior Features...")
financial_features = df_features.groupby('client').agg(
    total_money_in=('s_in', 'sum'),
    total_money_out=('s_out', 'sum'),
    avg_money_in=('s_in', 'mean'),
    avg_money_out=('s_out', 'mean'),
    max_money_in=('s_in', 'max'),
    max_money_out=('s_out', 'max'),
    std_money_in=('s_in', 'std'),
    std_money_out=('s_out', 'std')
).reset_index()

# Net flow (income - expenses)
financial_features['net_flow'] = (
    financial_features['total_money_in'] - financial_features['total_money_out']
)

# Money in/out ratio
financial_features['money_in_out_ratio'] = (
    financial_features['total_money_in'] / 
    (financial_features['total_money_out'] + 1)  # +1 to avoid division by zero
)

print(f"   ‚úì Created 10 financial features")

# ============================================================================
# GROUP 3: BALANCE FEATURES
# ============================================================================
print("\n3Ô∏è‚É£ Balance Features...")
balance_features = df_features.groupby('client').agg(
    final_balance=('bdt', 'last'),
    avg_balance=('bdt', 'mean'),
    min_balance=('bdt', 'min'),
    max_balance=('bdt', 'max'),
    std_balance=('bdt', 'std')
).reset_index()

# Balance trend (is it growing or declining?)
balance_features['balance_range'] = (
    balance_features['max_balance'] - balance_features['min_balance']
)

print(f"   ‚úì Created 6 balance features")

# ============================================================================
# GROUP 4: DEMOGRAPHIC FEATURES
# ============================================================================
print("\n4Ô∏è‚É£ Demographic Features...")
# Get unique client info
demo_features = df_features.groupby('client').agg(
    gender=('code_gender', 'first'),
    birth_place=('birth_place', 'first'),
    birthday=('birthday', 'first')
).reset_index()

# Calculate age
demo_features['birthday'] = pd.to_datetime(demo_features['birthday'])
demo_features['age'] = 2025 - demo_features['birthday'].dt.year

# Drop birthday (we don't need it anymore)
demo_features = demo_features.drop('birthday', axis=1)

print(f"   ‚úì Created: gender, birth_place, age")

# ============================================================================
# MERGE ALL FEATURES
# ============================================================================
print("\n5Ô∏è‚É£ Merging all features...")
df_final = transaction_freq.merge(financial_features, on='client', how='left')
df_final = df_final.merge(balance_features, on='client', how='left')
df_final = df_final.merge(demo_features, on='client', how='left')

# Add target
df_final = df_final.merge(client_target, on='client', how='inner')

print(f"\n‚úÖ FEATURE ENGINEERING COMPLETE!")
print(f"   Final dataset shape: {df_final.shape}")
print(f"   Features created: {df_final.shape[1] - 2} (excluding client ID and target)")
print(f"\nColumns: {list(df_final.columns)}")

# Show sample
print(f"\nüìä Sample of final dataset:")
print(df_final.head(3))

# Check for missing values in features
missing = df_final.isnull().sum()
if missing.sum() > 0:
    print(f"\n‚ö†Ô∏è Missing values in features:")
    print(missing[missing > 0])
else:
    print(f"\n‚úÖ No missing values in features!")
```

handling missing values again 

```python
# ============================================================================
# STEP 13: DATA PREPARATION FOR MODELING
# ============================================================================
print("\n" + "="*80)
print("STEP 13: PREPARING DATA FOR MODELING")
print("="*80)

# 13.1: Handle missing values in std columns
print("\n1Ô∏è‚É£ Handling missing values...")
print(f"   Missing values before:")
print(df_final[['std_money_in', 'std_money_out', 'std_balance']].isnull().sum())

# Fill missing std with 0 (means no variation - only 1 transaction)
df_final['std_money_in'] = df_final['std_money_in'].fillna(0)
df_final['std_money_out'] = df_final['std_money_out'].fillna(0)
df_final['std_balance'] = df_final['std_balance'].fillna(0)

print(f"\n   Missing values after:")
print(df_final[['std_money_in', 'std_money_out', 'std_balance']].isnull().sum())
print(f"   ‚úì All filled with 0!")

# 13.2: Handle categorical variables (birth_place)
print("\n2Ô∏è‚É£ Encoding categorical variables...")
print(f"   Unique birth places: {df_final['birth_place'].nunique()}")

# For birth_place, we'll use frequency encoding (too many categories for one-hot)
birth_place_freq = df_final['birth_place'].value_counts()
df_final['birth_place_freq'] = df_final['birth_place'].map(birth_place_freq)

# Drop original birth_place
df_final = df_final.drop('birth_place', axis=1)

print(f"   ‚úì Converted birth_place to frequency encoding")

# 13.3: Separate features and target
print("\n3Ô∏è‚É£ Separating features and target...")

# Drop client ID and target to get feature matrix
X = df_final.drop(['client', 'churn'], axis=1)
y = df_final['churn']

print(f"   Feature matrix (X): {X.shape}")
print(f"   Target vector (y): {y.shape}")
print(f"\n   Features: {list(X.columns)}")

# 13.4: Check final data
print("\n4Ô∏è‚É£ Final data check...")
print(f"   Any missing values: {X.isnull().sum().sum()}")
print(f"   Target distribution:")
print(y.value_counts())
print(f"   Target balance: {y.value_counts(normalize=True).round(3)*100}")

print("\n‚úÖ DATA READY FOR MODELING!")
print(f"   Total samples: {len(X)}")
print(f"   Total features: {X.shape[1]}")
```

SPLIOT&TRAIN

```python
# ============================================================================
# STEP 14: TRAIN-TEST SPLIT
# ============================================================================
print("\n" + "="*80)
print("STEP 14: TRAIN-TEST SPLIT")
print("="*80)

from sklearn.model_selection import train_test_split

# Split data: 70% train, 30% test
print("\n1Ô∏è‚É£ Splitting data...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nüìä SPLIT RESULTS:")
print(f"   Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"   Test set: {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")

print(f"\n   Train target distribution:")
print(y_train.value_counts())

print(f"\n   Test target distribution:")
print(y_test.value_counts())

print("\n‚úÖ TRAIN-TEST SPLIT COMPLETE!")
```

FEATURE SCALING

```python
# ============================================================================
# STEP 15: FEATURE SCALING
# ============================================================================
print("\n" + "="*80)
print("STEP 15: FEATURE SCALING")
print("="*80)

from sklearn.preprocessing import StandardScaler

print("\n1Ô∏è‚É£ Creating and fitting scaler...")
scaler = StandardScaler()

# Fit on training data only, then transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"   ‚úì Scaler fitted on training data")
print(f"   ‚úì Training data scaled: {X_train_scaled.shape}")
print(f"   ‚úì Test data scaled: {X_test_scaled.shape}")

print("\n2Ô∏è‚É£ Checking scaling results...")
print(f"   Before scaling - Train data mean: {X_train.mean().mean():.2f}")
print(f"   After scaling - Train data mean: {X_train_scaled.mean():.4f}")
print(f"   After scaling - Train data std: {X_train_scaled.std():.4f}")

print("\n‚úÖ FEATURE SCALING COMPLETE!")
print("\nüí° Why scaling? Makes features comparable and helps models converge faster!")
```

Testing

```python
# ============================================================================
# STEP 16: TRAIN LOGISTIC REGRESSION
# ============================================================================
print("\n" + "="*80)
print("STEP 16: LOGISTIC REGRESSION MODEL")
print("="*80)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("\n1Ô∏è‚É£ Training Logistic Regression model...")
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train_scaled, y_train)
print(f"   ‚úì Model trained!")

print("\n2Ô∏è‚É£ Making predictions on test set...")
y_pred_lr = lr_model.predict(X_test_scaled)
print(f"   ‚úì Predictions made for {len(y_pred_lr)} samples")

print("\n3Ô∏è‚É£ Evaluating model performance...")
lr_accuracy = accuracy_score(y_test, y_pred_lr)
print(f"\n   üéØ ACCURACY: {lr_accuracy:.4f} ({lr_accuracy*100:.2f}%)")

print("\n4Ô∏è‚É£ Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred_lr)
print(f"\n                Predicted")
print(f"              0 (Retained)  1 (Churned)")
print(f"Actual 0         {cm[0,0]}            {cm[0,1]}")
print(f"       1         {cm[1,0]}            {cm[1,1]}")

print("\n5Ô∏è‚É£ Detailed Classification Report:")
print(classification_report(y_test, y_pred_lr, target_names=['Retained', 'Churned']))

print("\n‚úÖ LOGISTIC REGRESSION COMPLETE!")
```

RANDOM FOREST

```python
# ============================================================================
# STEP 17: TRAIN RANDOM FOREST
# ============================================================================
print("\n" + "="*80)
print("STEP 17: RANDOM FOREST MODEL")
print("="*80)

from sklearn.ensemble import RandomForestClassifier

print("\n1Ô∏è‚É£ Training Random Forest model...")
rf_model = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    max_depth=10
)
rf_model.fit(X_train, y_train)  # RF doesn't need scaled data
print(f"   ‚úì Model trained with 100 trees!")

print("\n2Ô∏è‚É£ Making predictions on test set...")
y_pred_rf = rf_model.predict(X_test)
print(f"   ‚úì Predictions made for {len(y_pred_rf)} samples")

print("\n3Ô∏è‚É£ Evaluating model performance...")
rf_accuracy = accuracy_score(y_test, y_pred_rf)
print(f"\n   üéØ ACCURACY: {rf_accuracy:.4f} ({rf_accuracy*100:.2f}%)")

print("\n4Ô∏è‚É£ Confusion Matrix:")
cm_rf = confusion_matrix(y_test, y_pred_rf)
print(f"\n                Predicted")
print(f"              0 (Retained)  1 (Churned)")
print(f"Actual 0         {cm_rf[0,0]}            {cm_rf[0,1]}")
print(f"       1         {cm_rf[1,0]}            {cm_rf[1,1]}")

print("\n5Ô∏è‚É£ Detailed Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['Retained', 'Churned']))

print("\n‚úÖ RANDOM FOREST COMPLETE!")
```

```python
# ============================================================================
# STEP 18: MODEL COMPARISON
# ============================================================================
print("\n" + "="*80)
print("STEP 18: MODEL COMPARISON")
print("="*80)

# Create comparison table
comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [0.5612, 0.5180],
    'Accuracy %': ['56.12%', '51.80%']
})

print("\nüìä MODEL PERFORMANCE COMPARISON:")
print(comparison.to_string(index=False))

# Determine winner
print("\nüèÜ BEST MODEL: Logistic Regression")
print(f"   Better by: {(0.5612 - 0.5180)*100:.2f} percentage points")

print("\nüí° INSIGHT:")
print("   Logistic Regression performed better!")
print("   This is common with smaller datasets where")
print("   simpler models often outperform complex ones.")

print("\n‚úÖ MODEL COMPARISON COMPLETE!")
```

FEATURE IMPORTANCE

```python
# ============================================================================
# STEP 19: FEATURE IMPORTANCE ANALYSIS
# ============================================================================
print("\n" + "="*80)
print("STEP 19: FEATURE IMPORTANCE ANALYSIS")
print("="*80)

print("\n1Ô∏è‚É£ Extracting feature importance from Random Forest...")

# Get feature importance from Random Forest
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nüìä TOP 10 MOST IMPORTANT FEATURES:")
print(feature_importance.head(10).to_string(index=False))

print("\nüìä ALL FEATURES RANKED BY IMPORTANCE:")
print(feature_importance.to_string(index=False))

print("\n2Ô∏è‚É£ Key Insights:")
top_feature = feature_importance.iloc[0]
print(f"   ü•á Most important: {top_feature['Feature']} ({top_feature['Importance']:.4f})")

top_3 = feature_importance.head(3)['Feature'].tolist()
print(f"   üîù Top 3 features: {', '.join(top_3)}")

print("\n‚úÖ FEATURE IMPORTANCE ANALYSIS COMPLETE!")
```

```python
# ============================================================================
# STEP 19.5: INVESTIGATE AGE vs CHURN
# ============================================================================
print("\n" + "="*80)
print("INVESTIGATING: WHY IS AGE SO IMPORTANT?")
print("="*80)

# Merge age with target
age_analysis = df_final[['age', 'churn']].copy()

print("\n1Ô∏è‚É£ Age statistics by churn status:")
print("\nCHURNED clients (1):")
print(age_analysis[age_analysis['churn'] == 1]['age'].describe())

print("\nRETAINED clients (0):")
print(age_analysis[age_analysis['churn'] == 0]['age'].describe())

print("\n2Ô∏è‚É£ Average age comparison:")
churned_avg_age = age_analysis[age_analysis['churn'] == 1]['age'].mean()
retained_avg_age = age_analysis[age_analysis['churn'] == 0]['age'].mean()

print(f"   Churned clients average age: {churned_avg_age:.1f} years")
print(f"   Retained clients average age: {retained_avg_age:.1f} years")
print(f"   Difference: {abs(churned_avg_age - retained_avg_age):.1f} years")

if churned_avg_age > retained_avg_age:
    print("\nüí° INSIGHT: OLDER clients tend to churn more!")
else:
    print("\nüí° INSIGHT: YOUNGER clients tend to churn more!")

print("\n3Ô∏è‚É£ Age distribution by churn:")
age_bins = pd.cut(age_analysis['age'], bins=[0, 30, 40, 50, 60, 100], 
                   labels=['<30', '30-40', '40-50', '50-60', '60+'])
age_analysis['age_group'] = age_bins

churn_by_age = age_analysis.groupby('age_group')['churn'].agg(['sum', 'count', 'mean'])
churn_by_age.columns = ['Churned_Count', 'Total', 'Churn_Rate']
churn_by_age['Churn_Rate'] = churn_by_age['Churn_Rate'] * 100

print(churn_by_age)
```
